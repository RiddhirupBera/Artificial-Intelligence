```python
x = np.linspace(11,22,10)
y=x**2
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color="red")
```

```python
plt.xlim(11,28)
plt.ylim(11**2,22**2)

plt.plot(x,y,color="pink")
plt.xlabel("The x Axis")
plt.ylabel("The y Axis")
plt.title("The Title")

plt.show()
```

```python
#plt.xlim(11,22)
#plt.ylim(11**2,22**2)

plt.plot(x,y,color="pink",label="x vs y")
plt.plot(y,x,color="blue",label="y vs x")

plt.xlabel("The x Axis")
plt.ylabel("The y Axis")
plt.title("The Title")
plt.legend()

plt.show()
```

```python
courses = ["AI/ML","IOT","BlockChain","Python"]
```

```python
students = [400,657,543,865]
```

```python
plt.pie(students,labels=courses,autopct='%0.1f%%',colors=["purple","blue","pink","red"])
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.plot(x,y,color="green")
ax2.scatter(x,y,color="red")
ax3.bar(x,y,color="purple")
ax4.pie(students,labels=courses,autopct='%0.1f%%',colors=["purple","blue","pink","red"])
plt.show()
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
# data is linearly distributed and as total bill increases tip also increases
```

```python
sns.lineplot(x='size',y='tip',data=tips)
```

```python
sns.lineplot(x='size',y='tip',ci=0,data=tips) # ci--> confidence interval
```

```python
sns.lineplot(x='size',y='tip',ci=50,data=tips) # ci--> confidence interval
```

```python
sns.distplot(tips["tip"])
```

```python
sns.distplot(tips["tip"],bins=50)
```

```python
sns.barplot(x='time',y='tip',data=tips)
```

```python
sns.barplot(x='day',y='tip',data=tips)
```

```python
sns.boxplot(x='size',y='tip',data=tips)
```

```python
tips.describe() # this is visualized above
```

```python
sns.countplot(x="day",data=tips)
```

```python
sns.countplot(x="sex",data=tips)
```

```python
sns.catplot(x='smoker',y='tip',data=tips)
```

```python
sns.catplot(x='smoker',y='tip',hue='sex',data=tips)
```

```python
sns.catplot(x='day',y='tip',hue='sex',kind='bar',data=tips)
```
