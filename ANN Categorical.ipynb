```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer([("oh",OneHotEncoder(),[0])],remainder="passthrough")
y=ct.fit_transform(y).toarray()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
categorical = Sequential()
```

```python
categorical.add(Dense(units = 6,kernel_initializer = "random_uniform",activation="relu"))
categorical.add(Dense(units = 12,kernel_initializer = "random_uniform",activation="relu"))
categorical.add(Dense(units = 12,kernel_initializer = "random_uniform",activation="relu"))
categorical.add(Dense(units = 7,kernel_initializer = "random_uniform",activation="softmax"))
```

```python
categorical.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
history = categorical.fit(x_train,y_train,epochs=500,batch_size=32,validation_data=(x_test,y_test))
```

```python
# training accuracy high, testing low -> Overfitting
# training accuracy low, testing high -> Underfitting
# if both approx. equal --> Good
```

```python
categorical.add(Dense(units = 6,kernel_initializer = "random_uniform",activation="relu"))
categorical.add(Dropout(0.2))
categorical.add(Dense(units = 12,kernel_initializer = "random_uniform",activation="relu"))
categorical.add(Dropout(0.2))
categorical.add(Dense(units = 24,kernel_initializer = "random_uniform",activation="relu"))
categorical.add(Dropout(0.2))
categorical.add(Dense(units = 64,kernel_initializer = "random_uniform",activation="relu"))
categorical.add(Dropout(0.2))
categorical.add(Dense(units = 7,kernel_initializer = "random_uniform",activation="softmax"))
categorical.compile(optimizer="rmsprop",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
history = categorical.fit(x_train,y_train,epochs=1000,batch_size=100,validation_data=(x_test,y_test))
```

```python
categorical.add(Dense(units = 6,kernel_initializer = "random_uniform",activation="relu"))
categorical.add(Dense(units = 12,kernel_initializer = "random_uniform",activation="relu"))
categorical.add(Dense(units = 7,kernel_initializer = "random_uniform",activation="softmax"))
categorical.compile(optimizer="rmsprop",loss="categorical_crossentropy",metrics=["accuracy"])
history = categorical.fit(x_train,y_train,epochs=500,batch_size=32,validation_data=(x_test,y_test))
```
